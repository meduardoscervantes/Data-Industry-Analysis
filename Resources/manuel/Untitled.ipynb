```python
import pandas as pd
import gmaps
import os
```

```python
# Reas csv
map_df = pd.read_csv('resources/map_information')
data = pd.read_csv('resources/glassdoor_data_scientist_jobs.csv')
```

```python
new = []
for x in range(len(map_df)):
    new.append(f'{map_df["city"].tolist()[x]}, {map_df["state"].tolist()[x]}')
map_df["city_state"] = new
map_df
```

```python
# Create the figure
figure_layout = {
    'width': '900px',
    'height': '900px',
    'border': '1px solid black',
    'padding': '1px',
    'margin': '0 auto 0 auto'
}
fig = gmaps.figure(layout=figure_layout)
```

```python
counts = []
for x in map_df["city_state"].tolist():
    counts.append(len(data.loc[data["Location"] == x]))
counts
```

```python
lat_long = map_df[["lat","long"]]
heat_layer = gmaps.heatmap_layer(lat_long, weights=counts)
fig.add_layer(heat_layer)
fig
```
